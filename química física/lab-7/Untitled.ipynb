```python
!pip install matplotlib

import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Определяем функцию, описывающую дифференциальное уравнение
def model(y, t):
    dydt = -2 * y
    return dydt

# Начальные условия
y0 = 1  # начальное значение y
t = np.linspace(0, 5, 100)  # временной интервал от 0 до 5 с 100 точками

# Решаем ОДУ
solution = odeint(model, y0, t)

# Визуализация результата
plt.plot(t, solution)
plt.title('Решение ОДУ: dy/dt = -2y')
plt.xlabel('Время t')
plt.ylabel('y(t)')
plt.grid()
plt.show()
```
